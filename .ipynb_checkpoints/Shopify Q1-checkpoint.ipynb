# Problem Definition

On Shopify, we have exactly 100 sneaker shops, and each of these shops sells only one model of shoe. We want to do some analysis of the average order value (AOV). When we look at orders data over a 30 day window, we naively calculate an AOV of $3145.13. Given that we know these shops are selling sneakers, a relatively affordable item, something seems wrong with our analysis. 

## Questions
1: Think about what could be going wrong with our calculation. Think about a better way to evaluate this data. 

2: What metric would you report for this dataset?

3: What is its value?


```python
import pandas as pd


aov_predicted = 3145.13
aov_mean = 0
aov_mode = 0
aov_median = 0
```

```python
# CSV read into Pandas dataFrame
order_data = pd.read_csv("2019 Winter Data Science Intern Challenge Data Set - Sheet1.csv")
order_data.head(10)
```

```python
def calcMeanAOV():
    price_total = 0
    # Increment through each row in the data structure
    for x in range(len(order_data)):
        # Sum the order_amount from each row
        price_total += order_data["order_amount"][x]
        
    # Return the mean average value
    return price_total / len(order_data)    
```

```python
aov_mean = calcMeanAOV(order_data)
```

```python
# Reordering the data in decending order of 'order_amount'
order_data_sorted = order_data.sort_values(by=["order_amount"], ascending=False)
order_data_sorted.head(10)
```

```python
def calcMedianAOV():
    if(len(order_data) % 2 == 0):
        print(order_data["order_amount"][int(len(order_data)/2)])
    else:
        print(order_data["order_amount"][int(len(order_data)/2)])
```
